```python
"""1. Given a sentence as txt, return True if any two adjacent words have this
property: One word ends with a vowel, while the word immediately after
begins with a vowel (a e i o u).
Examples
vowel_links("a very large appliance") ➞ True
vowel_links("go to edabit") ➞ True
vowel_links("an open fire") ➞ False
vowel_links("a sudden applause") ➞ False"""

def vowel_links(n):
    v = "aeiou"
    l = []
    k = 0
    s = n.split()
    for i in s:
        l.append(len(i))
    for j in l:
        lst = (s[k][j-1])
        k += 1
        if lst in v:
            if k == len(l):
                return False
            srt = (s[k][0])
            if srt in v:
                return True
        else:
            continue
    return False

print(vowel_links("a very large appliance"))
print(vowel_links("go to edabit"))
print(vowel_links("an open fire"))
print(vowel_links("a sudden applause"))
```

```python
"""2. You are given three inputs: a string, one letter, and a second letter.
Write a function that returns True if every instance of the first letter occurs
before every instance of the second letter.
Examples
first_before_second("a rabbit jumps joyfully", "a", "j") ➞ True
# Every instance of "a" occurs before every instance of "j".
first_before_second("knaves knew about waterfalls", "k", "w") ➞ True
first_before_second("happy birthday", "a", "y") ➞ False
# The "a" in "birthday" occurs after the "y" in "happy".
first_before_second("precarious kangaroos", "k", "a") ➞ False"""

#short method
def first_before_second(s, l1, l2):
    fst_lett = 0
    lst = [l1, l2]
    lettr = []
    str_tot = len(s)
    for i in lst:
        for j in range(str_tot):
            if i == s[j]:
                fst_lett = j
                lettr.append(fst_lett)
                break
    if lettr[0] < lettr[1]:
        if l1 in s[lettr[1]:]:
            return False
        else:
            return True
    else:
        return False

print(first_before_second("a rabbit jumps joyfully", "a", "j"))
print(first_before_second("knaves knew about waterfalls", "k", "w"))
print(first_before_second("happy birthday", "a", "y"))
print(first_before_second("precarious kangaroos", "k", "a"))

# #Long Method
# def first_before_second(s, l1, l2):
#     fst_lett = 0
#     sec_lett = 0
#     lettr = []
#     lst = [l1, l2]
#     str_tot = len(s)
#     for i in lst:
#         for j in range(str_tot):
#             if i == s[j]:
#                 fst_lett = j
#                 lettr.append(fst_lett)
#                 break
#     if lettr[0] < lettr[1]:
#         for k in l1:
#             for m in range(str_tot):
#                 if k == s[m]:
#                     sec_lett = m
#         if sec_lett < lettr[1]:
#             return True
#         else:
#             return False
#     else:
#         return False
#
# print(first_before_second("a rabbit jumps joyfully", "a", "j"))
# print(first_before_second("knaves knew about waterfalls", "k", "w"))
# print(first_before_second("happy birthday", "a", "y"))
# print(first_before_second("precarious kangaroos", "k", "a"))
```

```python
"""3. Create a function that returns the characters from a list or string r on odd or
even positions, depending on the specifier s. The specifier will be "odd" for
items on odd positions (1, 3, 5, ...) and "even" for items on even positions (2,4, 6, ...).
Examples
char_at_pos([2, 4, 6, 8, 10], "even") ➞ [4, 8]
# 4 & 8 occupy the 2nd & 4th positions
char_at_pos("EDABIT", "odd") ➞ "EAI"
# "E", "A" and "I" occupy the 1st, 3rd and 5th positions
char_at_pos(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"], "odd") ➞ ["A", "B", "T", "A", "I", "Y"]"""

def char_at_pos(n, s):
    result = []
    stri = ""
    for i in range(1, len(n)+1):
        if s =="even":
            if i % 2 == 0:
                result.append(n[i-1])
        if s == "odd":
            if i % 2 != 0:
                if type(n) == str:
                    stri = stri + n[i-1]
                else:
                    result.append(n[i-1])
    if type(n) == str:
        return stri
    else:
        return result

print(char_at_pos([2, 4, 6, 8, 10], "even"))
print(char_at_pos("EDABIT", "odd"))
print(char_at_pos(["A", "R", "B", "I", "T", "R", "A", "R", "I", "L", "Y"], "odd"))

```

```python
"""4. Write a function that returns the greatest common divisor of all list
elements. If the greatest common divisor is 1, return 1.
Examples
GCD([10, 20, 40]) ➞ 10
GCD([1, 2, 3, 100]) ➞ 1
GCD([1024, 192, 2048, 512]) ➞ 64"""

def GCD(n):
    i = 0
    gcd = 0
    hcf1 = n[i]
    while i < len(n):
        hcf2 = n[i + 1]
        if hcf1 > hcf2:
            max = hcf1
        else:
            max = hcf2
        for j in range(1, max + 1):
            if (hcf1 % j == 0) and (hcf2 % j == 0):
                gcd = j
        hcf1 = gcd
        if hcf2 == n[len(n)-1]:
            return gcd
        i = i + 1

print(GCD([128,96]))
print(GCD([10, 20, 40]))
print(GCD([1, 2, 3, 100]))
print(GCD([1024, 192, 2048, 512]))

```

```python
"""5. A number/string is a palindrome if the digits/characters are the same when
read both forward and backward. Examples include "racecar" and 12321.
Given a positive number n, check if n or the binary representation of n is
palindromic. Return the following:
- "Decimal only." if only n is a palindrome.
- "Binary only." if only the binary representation of n is a palindrome.
- "Decimal and binary." if both are palindromes.
- "Neither!" if neither are palindromes.
Examples
palindrome_type(1306031) ➞ "Decimal only."
# decimal = 1306031
# binary = "100111110110110101111"
palindrome_type(427787) ➞ "Binary only."
# decimal = 427787
# binary = "1101000011100001011"
palindrome_type(313) ➞ "Decimal and binary."
# decimal = 313
# binary = 100111001
palindrome_type(934) ➞ "Neither!"
# decimal = 934
# binary = "1110100110" """

## Using Binary Function
def palindrome_type(n):
    flag = 0
    dec_num = str(n)
    n_rev = dec_num[::-1]
    if n_rev == dec_num:
        flag = 1

    f = 0
    binary = (bin(n)[2:])
    rev_bin = "".join(reversed(binary))
    if binary == rev_bin:
        f = 1
    if flag == 1 and f == 1:
        return "Decimal and binary"
    elif flag == 1 and f == 0:
        return "Decimal Only"
    elif flag == 0 and f == 1:
        return "Binary Only"
    else:
        return "Neither!"

print(palindrome_type(1306031))
print(palindrome_type(427787))
print((palindrome_type(313)))
print(palindrome_type(934))


## without using bin() function
# def palindrome_type(n):
#     flag = 0
#     dec_num = str(n)
#     n_rev = dec_num[::-1]
#     if n_rev == dec_num:
#         flag = 1
#
#     f = 0
#     r = ""
#     while n > 0:
#         rem = n % 2
#         r = r + str(rem)
#         q = n // 2
#         n = q
#     bin_rev =  "".join(reversed(r))
#     if r == bin_rev:
#         f = 1
#     if flag == 1 and f == 1:
#         return "Decimal and binary"
#     elif flag == 1 and f == 0:
#         return "Decimal Only"
#     elif flag == 0 and f == 1:
#         return "Binary Only"
#     else:
#         return "Neither!"
#
# print(palindrome_type(1306031))
# print(palindrome_type(427787))
# print((palindrome_type(313)))
# print(palindrome_type(934))
```
