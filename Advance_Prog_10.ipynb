```python
"""2. Write three functions:
1. boolean_and
2. boolean_or
3. boolean_xor
These functions should evaluate a list of True and False values, starting from
the leftmost element and evaluating pairwise.
Examples
boolean_and([True, True, False, True]) ➞ False
# [True, True, False, True] => [True, False, True] => [False, True] => False
boolean_or([True, True, False, False]) ➞ True
# [True, True, False, True] => [True, False, False] => [True, False] => True
boolean_xor([True, True, False, False]) ➞ False
# [True, True, False, False] => [False, False, False] => [False, False] => False"""

def boolean_and(n):
    print("boolean_and:")
    and_out = True
    for i in n:
        if i == and_out:
            and_out = True
        else:
            and_out = False
    return and_out

def boolean_or(m):
    print("boolean_or:")
    out = True
    for i in m:
        out = i or out
    return out

def boolean_xor(b):
    print("boolean_xor:")
    out = 0
    i = 0
    while i < len(b):
        if i == 0:
            out = b[i] ^ b[i+1]
            i = i + 2
        else:
            out = b[i] ^ out
            i += 1
    return out

print(boolean_and([True, True, False, True]))
print(boolean_or([True, True, False, False]))
print(boolean_xor([True, True, False, False]))
```

```python
"""3. Create a function that creates a box based on dimension n.
Examples
make_box(5) ➞ [
"#####",
"# #",
"# #",
"# #",
"#####"
]
make_box(3) ➞ [
"###",
"# #",
"###"
]
make_box(2) ➞ [
"##",
"##"
]
make_box(1) ➞ [
"#"
]"""

def make_box(n):
    print("Output: ", n)
    new_list = []
    for row in range(n):
        col_list = []
        for column in range(n):
            if row == 0 or row == n-1:
                col_list.append("#")
            elif column == 0 or column == n-1:
                col_list.append("#")
            else:
                col_list.append(" ")
        new_list.append(col_list)

    for sub_list in new_list:
        for element in sub_list:
            print(element, end="")
        print()

make_box(5)
make_box(3)
make_box(2)
make_box(1)
```

```python
"""4. Given a common phrase, return False if any individual word in the phrase
contains duplicate letters. Return True otherwise.
Examples
no_duplicate_letters("Fortune favours the bold.") ➞ True
no_duplicate_letters("You can lead a horse to water, but you can't make him drink.") ➞ True
no_duplicate_letters("Look before you leap.") ➞ False
# Duplicate letters in "Look" and "before".
no_duplicate_letters("An apple a day keeps the doctor away.") ➞ False
# Duplicate letters in "apple", "keeps", "doctor", and "away"."""

def no_duplicate_letters(sentes):
    sp = sentes.split()
    l1 = []
    for i in range(len(sp)):
        for j in range(len(sp[i])):
            k = sp[i][j]
            c = sp[i].count(k)
            if c >= 2:
                l1.append(sp[i])
                break
    if l1 == []:
        return True
    else:
        print(l1, end=": ")
        return False

print(no_duplicate_letters("Fortune favours the bold."))
print(no_duplicate_letters("You can lead a horse to water, but you can't make him drink."))
print(no_duplicate_letters("Look before you leap."))
print(no_duplicate_letters("An apple a day keeps the doctor away."))
```
